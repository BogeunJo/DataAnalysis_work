# 1. 인구 피라미드

## 남자 데이터 정의

```python
import pandas as pd
df_m = pd.read_excel('202305_202305_연령별인구현황_월간_1세단위.xlsx', skiprows=3, index_col='행정기관', usecols='B, E:DA')
df_m.head(3)
```

```python
pd.set_option('display.max_columns', 210)
pd.set_option('display.max_seq_items', None)
pd.set_option('display.max_rows', None)
```

```python
df_m.iloc[0]
```

```python
df_m.iloc[0,1]
```

```python
df_m.iloc[0] = df_m.iloc[0].astype(str).apply(lambda x: x if ',' not in x and int(x) <= 1000 else x.replace(',', '')).astype(int)
```

```python
df_m
```

## 여자 데이터 정의

```python
df_w = pd.read_excel('202305_202305_연령별인구현황_월간_1세단위.xlsx', skiprows=3, index_col='행정기관', usecols='B, DD:GZ')
df_w.head(3)
```

```python
df_w.iloc[0]
```

```python
df_w.columns=df_m.columns # 컬럼명 통일
df_w.columns
```

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(',','').astype(int)
```

```python
df_w
```

## 데이터 시각화

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.size'] = 15 
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df_m.columns
```

```python
df_w.columns
```

```python
plt.figure(figsize=(10,7))
plt.barh(df_m.columns, -df_m.iloc[0]//1000) # 단위 : 천 명
plt.barh(df_w.columns, df_w.iloc[0]//1000)
plt.yticks(range(0, len(df_m.columns), 5), df_m.columns[::5])
plt.title('2023년 대한민국 인구 피라미드_1세 단위')
plt.savefig('2023_인구피라미드_1세 단위.png', dpi=100)
plt.show()
```
