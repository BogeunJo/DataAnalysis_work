```python
df = df.T
df
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.size'] = 15 
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
plt.plot(df.index, df['출생아 수'])
plt.plot(df.index, df['합계 출산율'])
```

```python
fig, ax1=plt.subplots(figsize=(10,7))
ax1.plot(df.index, df['출생아 수'], color='#ff812d')

ax2= ax1.twinx() # x 축을 공유하는 쌍둥이 axis
ax2.plot(df.index, df['합계 출산율'], color='#ffd100')
```

```python
fig, ax1=plt.subplots(figsize=(8,4))
ax1.set_ylabel('출생아 수 (천 명)')
ax1.set_ylim(250,600)
ax1.set_yticks([300,400,500,600])
ax1.bar(df.index, df['출생아 수'], color='#ff812d')

ax2= ax1.twinx() # x 축을 공유하는 쌍둥이 axis
ax2.set_ylabel('합계 출산율 (가임 여성 1명당 명)')
ax2.set_ylim(0,1.3)
ax2.set_yticks([0,1])
ax2.plot(df.index, df['합계 출산율'], color='#ffd100')
```

```python
fig, ax1=plt.subplots(figsize=(8,4))
ax1.set_ylabel('출생아 수 (천 명)')
ax1.set_ylim(250,600)
ax1.set_yticks([300,400,500,600])
ax1.bar(df.index, df['출생아 수'], color='#ff812d')
for idx, val in enumerate(df['출생아 수']):
    ax1.text(idx, val + 7, val, ha='center')

ax2= ax1.twinx() # x 축을 공유하는 쌍둥이 axis
ax2.set_ylabel('합계 출산율 (가임 여성 1명당 명)')
ax2.set_ylim(0,1.3)
ax2.set_yticks([0,1])
ax2.plot(df.index, df['합계 출산율'], color='#ffd100')
```

```python
fig, ax1=plt.subplots(figsize=(8,4))
fig.suptitle('출생아 수 및 합계 출산율 (2012~2021)')
# ax1.set_ylabel('출생아 수 (천 명)')
ax1.set_ylabel('출생아 수 (천 명)', rotation=0, ha='right', va='center')
ax1.yaxis.set_label_coords(-0.1, 0.5)
ax1.set_ylim(250,600)
ax1.set_yticks([300,400,500,600])
ax1.bar(df.index, df['출생아 수'], color='#03a9f5')
for idx, val in enumerate(df['출생아 수']):
    ax1.text(idx, val + 7, val, ha='center')

ax2= ax1.twinx() # x 축을 공유하는 쌍둥이 axis
ax2.set_ylabel('합계 출산율 (가임 여성 1명당 명)')
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
ax2.plot(df.index, df['합계 출산율'], color='#ffd100', marker='o', ms=9, lw=3, mfc='w', mew=3)
for idx, val in enumerate(df['합계 출산율']):
    ax2.text(idx, val + 0.08, val, ha='center')
```
